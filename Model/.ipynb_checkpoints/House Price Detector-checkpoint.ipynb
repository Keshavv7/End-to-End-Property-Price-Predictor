```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)
```

```python
data = pd.read_csv('Bengaluru_House_Data.csv')
data
```

```python
data.shape
```

<b> Reviewing the features </b>

```python
# Let's first check the number of unique values in area_type feature
data.groupby('area_type')['area_type'].agg('count')
```

```python
# For now let us drop some columns
data2 = data.drop(['area_type', 'availability', 'society'], axis='columns')
data2
```

<b> Data Cleaning and Processing </b>

```python
#Handling NA values
data2.isnull().sum()
```

```python
data2['balcony'] = data2['balcony'].fillna(0) #Filling NA balcony values with 0
#Dropping rest of the rows with NA values
data3 = data2.dropna()
data3.isnull().sum()
```

```python
data3
```

```python
# We can see that the size column is very consistent in data representation
data3['size'].unique()
```

```python
# We can convert Bedroom to BHK or just simply keep count of bedrooms as integer
data3['bedrooms'] = data3['size'].apply(lambda x: int(x.split(' ')[0]))
data3
```

```python
data4 = data3.drop('size', axis='columns')
data4
```

```python
data4['bedrooms'].unique()
```

```python
# We can see there exist houses with 43 bedrooms too, which is unusual.
data4[data4.bedrooms > 20]
```

```python
# Clearly 2400 sqft house cannot have 43 bedrooms
data4.total_sqft.unique()
```

```python
# We can see the value '1133-1384' which is a range and not an actual value
# We can take mean of these numbers. 
# Let us examine the total_sqft field more
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

data4[~data4['total_sqft'].apply(is_float)].head(20)
```

```python
# We can see there are all kinds of values under sqft
# One way is to ignore all those rows with values other than ranges
def convert_to_num(x):
    words = x.split('-')
    if (len(words)==2):
        return (float(words[0])+float(words[1]))/2 #Convert range to avg
    try:
        return float(x) #Convert string nums to float
    except:
        return None # Discard everything else

data5 = data4.copy()
data5['total_sqft'] = data4['total_sqft'].apply(convert_to_num)
data5.head(20)
```

<b> Feature Engineering </b>

```python
# We know that price per sqft is an important feature
data6 = data5.copy()
data6['price_per_sqft'] = data6['price']*100000/data6['total_sqft']
data6
```

```python
len(data6.location.unique())
```

```python
#Too many categories
data6.location = data6.location.apply(lambda x: x.strip()) #Removing trailing/leading spaces
location_stats = data6.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
len(data6.location.unique())
```

```python
data6.location = data6.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(data6.location.unique())
```

```python
# Now the one-hot encoding will only have around 250 columns

# Let us now remove outliers and abnormal data points
```

We may recall that there were several houses with few bedrooms and very large size or very high number of bedrooms and very low size. We can keep a rough estimate of a normal bedroom size ( should be around 200sqft to 600 sqft or so ). Any data points which violate this range are abnormal. 

```python
data6[data6.total_sqft/data6.bedrooms<300]
```

```python
# We can see that all these rows have unusual values
data6.shape
```

```python
data7 = data6[~(data6.total_sqft/data6.bedrooms<300)]
data7
```

```python
# Let us check price per sqft
data7.price_per_sqft.describe()
```

```python
# We can see that the min and max values are too extreme
# If data follows normal distribution, then around 70% of data points should within the standard deviation of mean
def handle_extreme_price(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df], ignore_index=True)
    return df_out
final = handle_extreme_price(data7)
final
```

```python
# Examining the csv file also reveals that for same location, some rows are such that price for 2 bedrooms is higher than that of 3 bedrooms.
#  rows might affect our model. 
#Let us visualize this

def scatter_plotter(df, location):
    bhk2 = df[(df.location==location) & (df.bedrooms==2)]
    bhk3 = df[(df.location==location) & (df.bedrooms==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft, bhk2.price,color='blue',label='2BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price,marker='*',color='red',label='3BHK', s=50)
    plt.xlabel('Total Square Feet Area')
    plt.ylabel('Price')
    plt.title(location)
    plt.legend()
```

```python
scatter_plotter(final, 'Rajaji Nagar')
```

```python
#We can see around 1750sqft, 2 bedroom price is higher than 3 bedrooms
# Need to remove these outliers
```

```python
scatter_plotter(final, 'Hebbal')
```

One normal way is to check the mean and std for 1 bedroom apartments in one area, and then make sure that the 2 bedroom stats are above these vaues. 

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bedrooms'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bedrooms'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

final2 = remove_bhk_outliers(final)
final2
```

```python
final3 = final2[final2.bath < final2.bedrooms+2]
final3
```

```python
final4 = final3.drop(['price_per_sqft'], axis='columns')
final4
```

```python
# We shall now convert categorical text feature (location) into one hot encoded variable

dummies = pd.get_dummies(final4.location)
dummies
```

```python
#Note that we can drop one column, so we can drop others
final5 = pd.concat([final4, dummies.drop('other',axis='columns')], axis='columns')
final5
```

<b> Now we are ready to create the training and testing sets </b> 

```python
X = final5.drop(['location', 'price'], axis='columns')
y = final5['price']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
X_train
```

```python
X_test
```

<b> Creating the model </b>

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
```

Performing K-Fold Cross Validation

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

Let us try some other regression models using GridSearchCV

```python
# Model selection and Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model(X,y):
    algorithms = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
                'fit_intercept' : [True, False],
                'positive' : [True, False]
            }
        },
        'lasso' : {
            'model' : Lasso(),
            'params' : {
                'alpha' : [1,2],
                'selection' : ['random', 'cyclic']
            }
        },
        'decision_tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['mse', 'friedman_mse'],
                'splitter' : ['best', 'random']
            }
        }
    }
    
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algorithms.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model' : algo_name,
            'best_score' : gs.best_score_,
            'best_params' : gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model','best_score','best_params'])


results = find_best_model(X, y)
results
```

```python
# We can see that the best model is Linear Regression
# Now we can predict

regression_model = LinearRegression(fit_intercept=False, positive=False)
regression_model.fit(X_train.values, y_train.values)

def predict_price(location, sqft, bath, balcony, bedrooms):
    loc_index = np.where(X.columns==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = balcony
    x[3] = bedrooms
    if loc_index >= 0:
        x[loc_index] = 1
        
    return regression_model.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar', 1000, 2, 1, 2)
```

```python
place = 'Indira Nagar'
size_in_sqft = 2000
bedrooms = 3
bathrooms = 3
balcony = 2
price = (predict_price(place, size_in_sqft, bathrooms, balcony, bedrooms))
if (price < 100):
    print(int(price), 'lakhs rupees')
else:
    ans = '{:.2f}'.format(price/100)
    print(ans, 'crore rupees')
```
